# 5. Categorizing and Tagging Words

Word classes ,e.g. nouns, verbs, adjectives, and adverbs, are useful categories for many language processing tasks.

The process of classifying words into their parts of speech and labeling them accordingly is known as **part-of-speech tagging**, **POS-tagging**, or simply **tagging**. 

Parts of speech are also known as **word classes** or **lexical categories**.

**POS tagging** is the second step in the typical NLP pipeline, following **tokenization**.

# 1.   Using a Tagger

A part-of-speech tagger, or **POS-tagger**, processes a sequence of words, and attaches a part of speech tag to each word.

```python
import nltk
from nltk import word_tokenize
```

```python
text = word_tokenize("And now for something completely different")
```

```python
text
```

```python
nltk.pos_tag(text)
```

- and is **CC**, a coordinating conjunction;

- now and completely are **RB**, or adverbs;

- for is **IN**, a preposition;

- something is **NN**, a noun;

- and different is **JJ**, an adjective.

```python
import jieba
import jieba.posseg as pseg
```

```python
# jieba的词性标签含义见: https://github.com/fxsjy/jieba
# 4.词性标注 部分
for word, pos in pseg.cut('坚持把解决好“三农”问题作为全党工作重中之重 举全党全社会之力推动乡村振兴'):
    print(word, pos)
```

Let's look at another example, this time including some **homonyms** (同形异义，如refuse, permit).

```python
# 他们拒绝让我们获得垃圾许可证
text = word_tokenize("They refuse to permit us to obtain the refuse permit")
```

```python
nltk.pos_tag(text)
```

*refUSE* is a verb meaning "deny," while *REFuse* is a noun meaning "trash" (i.e. they are not homophones, 名词重音在前，动词重音在后).

permit动词读作 /pə'mɪt/，名词读作 /'pɜːmɪt/.

## 2.6   Adjectives and Adverbs

**Adjectives describe nouns**, and can be used as **modifiers** (e.g. large in the large pizza), or **in predicates** (e.g. the pizza is large).

**Adverbs modify verbs** to specify the time, manner, place or direction of the event described by the verb (e.g. quickly in the stocks fell quickly). Adverbs may also **modify adjectives** (e.g. really in Mary's teacher was really nice).

English has several categories of **closed class words** in addition to **prepositions**, such as **articles** (also often called **determiners** 限定词) (e.g., the, a), **modals** (e.g., should, may), and **personal pronouns** (e.g., she, they).

## 2.7   Unsimplified Tags

Let's find the most frequent nouns of each noun part-of-speech type.

```python
# 返回语料库中给定词性的前5个高频词
def findtags(tag_prefix, tagged_text):
    cfd = nltk.ConditionalFreqDist((tag, word) for (word, tag) in tagged_text
                                  if tag.startswith(tag_prefix))
    return dict((tag, cfd[tag].most_common(5)) for tag in cfd.conditions())
```

```python
tagdict = findtags('NN', nltk.corpus.brown.tagged_words(categories='news'))
```

```python
# 返回各类名词中，词频排名前五的单词
for tag in sorted(tagdict):
    print(tag, tagdict[tag])
```

There are many variants of NN, the most important contain:

$ for possessive nouns,

S for plural nouns (since plural nouns typically end in s) 

P for proper nouns.

In addition, most of the tags have suffix modifiers:

-NC for citations,

-HL for words in headlines,

-TL for titles

Next, let's look at some larger context, and find words **involving particular sequences of tags and words** (in this case `<Verb> to <Verb>`).

```python
# 寻找符合特定模式的文本片段

def process(sentence):
    for (w1,t1), (w2,t2), (w3,t3) in nltk.trigrams(sentence):
        if (t1.startswith('V') and t2 == 'TO' and t3.startswith('V')):
            print(w1, w2, w3)
```

```python
from nltk.corpus import brown

for tagged_sent in brown.tagged_sents():
    process(tagged_sent)
```

Finally, let's look for words that are highly ambiguous as to their part of speech tag.

```python
# 看一下在词性标签上存在歧义的词语

brown_news_tagged = brown.tagged_words(categories='news', tagset='universal')
data = nltk.ConditionalFreqDist((word.lower(), tag) for (word, tag) in brown_news_tagged)
```

```python
# word是条件，tag是事件
for word in sorted(data.conditions()):
    if len(data[word]) > 3:
        tags = [tag for (tag, _) in data[word].most_common()]
        print(word, ' '.join(tags))
```

# 4.   Automatic Tagging

We will see that the tag of a word depends on **the word** and **its context within a sentence**.

For this reason, we will be working with **data at the level of (tagged) sentences rather than words**.

```python
from nltk.corpus import brown
```

```python
brown_tagged_sents = brown.tagged_sents(categories='news')
```

```python
brown_sents = brown.sents(categories='news')
```

## 4.1   The Default Tagger

```python
tags = [tag for (word, tag) in brown.tagged_words(categories='news')]
```

```python
# 标注数据集中最常见的标签是NN
nltk.FreqDist(tags).max()
```

Now we can create a tagger that tags everything as NN. It establishes **an important baseline for tagger performance**. 

```python
raw = 'I do not like green eggs and ham, I do not like them Sam I am!'
```

```python
tokens = nltk.word_tokenize(raw)
```

```python
default_tagger = nltk.DefaultTagger('NN')
```

```python
default_tagger.tag(tokens)
```

Unsurprisingly, this method performs rather poorly. On a typical corpus, it will **tag only about an eighth of the tokens correctly**, as we see below:

```python
default_tagger.evaluate(brown_tagged_sents)
```

## 4.2   The Regular Expression Tagger

The regular expression tagger assigns tags to tokens on the basis of **matching patterns**. 

For instance, we might guess that any word ending in ed is the past participle of a verb, and any word ending with 's is a possessive noun.

We can express these as a list of regular expressions:

```python
patterns = [
     (r'.*ing$', 'VBG'),                # gerunds
     (r'.*ed$', 'VBD'),                 # simple past
     (r'.*es$', 'VBZ'),                 # 3rd singular present
     (r'.*ould$', 'MD'),                # modals
     (r'.*\'s$', 'NN$'),                # possessive nouns
     (r'.*s$', 'NNS'),                  # plural nouns
     (r'^-?[0-9]+(\.[0-9]+)?$', 'CD'),  # cardinal numbers
     (r'.*', 'NN')                      # nouns (default)
]
```

Note that these are processed **in order**, and the first one that matches is applied.

```python
regexp_tagger = nltk.RegexpTagger(patterns)
```

```python
print(regexp_tagger.tag(brown_sents[3]))
```

```python
regexp_tagger.evaluate(brown_tagged_sents)
```

Now its right about a fifth of the time. 相比全部标注为名词，准确率上升了7%。

## 4.3   The Lookup Tagger

A lot of high-frequency words do not have the NN tag.

Let's find **the hundred most frequent words** and **store their most likely tag**.

```python
fd = nltk.FreqDist(brown.words(categories='news'))
```

```python
cfd = nltk.ConditionalFreqDist(brown.tagged_words(categories='news'))
```

```python
# 最高频的100个词
most_freq_words = fd.most_common(100)
```

```python
# 为每个高频词找出其最高频的词性标记
likely_tags = dict((word, cfd[word].max()) for (word, _) in most_freq_words)
```

```python
print(likely_tags.items())
```

```python
baseline_tagger = nltk.UnigramTagger(model=likely_tags)
```

```python
# 此时，我们已经把baseline的准确率提高到了45%，比正则表达式分词器提高了25%
baseline_tagger.evaluate(brown_tagged_sents)
```

Let's see what it does on some untagged input text:

```python
sent = brown.sents(categories='news')[3]
```

```python
print(sent)
```

```python
print(baseline_tagger.tag(sent))
```

Many words have been assigned a tag of None, because they were not among the 100 most frequent words. 

In these cases we would like to **assign the default tag of NN**. 

In other words, we want to **use the lookup table first**, and if it is unable to assign a tag, then use the default tagger, a process known as **backoff** (退避).

```python
baseline_tagger = nltk.UnigramTagger(model=likely_tags,backoff=nltk.DefaultTagger('NN'))
```

```python
# 评估lookup tagger + default tagger的准确率
def performance(cfd, wordlist):
    lt = dict((word, cfd[word].max()) for word in wordlist)
    baseline_tagger = nltk.UnigramTagger(model=lt, backoff=nltk.DefaultTagger('NN'))
    return baseline_tagger.evaluate(brown.tagged_sents(categories='news'))
```

```python
import pylab
word_freqs = nltk.FreqDist(brown.words(categories='news')).most_common()
words_by_freq = [w for (w, _) in word_freqs]
cfd = nltk.ConditionalFreqDist(brown.tagged_words(categories='news'))
sizes = 2 ** pylab.arange(15)
perfs = [performance(cfd, words_by_freq[:size]) for size in sizes]
pylab.plot(sizes, perfs, '-bo')
pylab.title('Lookup Tagger Performance with Varying Model Size')
pylab.xlabel('Model Size')
pylab.ylabel('Performance')
pylab.show()
```

观察到性能最初随着模型尺寸的增长而快速增长，最终达到稳定状态（model size > 1024），此时模型尺寸的大幅增加几乎不会带来性能的改善。

```python
sizes
```

lookup tagger保存的词越多，准确率越高

```python
performance(cfd, words_by_freq[:100])
```

组合的tagger把baseline的准确率提升为58%，比lookup tagger提高了13%。

# 5.   N-Gram Tagging

## 5.1   Unigram Tagging


Unigram taggers are based on a simple statistical algorithm: for each token, **assign the tag that is most likely for that particular token**. 

For example, it will assign the tag JJ to any occurrence of the word *frequent*, since frequent is used as an adjective more often than it is used as a verb. 

**A unigram tagger behaves just like a lookup tagger**.

```python
from nltk.corpus import brown
```

```python
brown_tagged_sents = brown.tagged_sents(categories='news')
```

```python
brown_sents = brown.sents(categories='news')
```

```python
unigram_tagger = nltk.UnigramTagger(brown_tagged_sents)
```

```python
print(unigram_tagger.tag(brown_sents[2007]))
```

```python
unigram_tagger.evaluate(brown_tagged_sents)
```

We **train** a UnigramTagger by **specifying tagged sentence data as a parameter** when we initialize the tagger. 

The training process involves inspecting the tag of each word and storing the most likely tag for any word in a dictionary.

## 5.2   Separating the Training and Testing Data

Now that we are **training a tagger on some data**, we must be careful **not to test it on the same data**, as we did in the above example. 

A tagger that **simply memorized its training data** and made no attempt to construct **a general model** would get a perfect score, but would also be **useless for tagging new text**. 

Instead, we should split the data, **training on 90%** and **testing on the remaining 10%**.

```python
size = int(len(brown_tagged_sents) * 0.9)
```

```python
size
```

```python
train_sents = brown_tagged_sents[:size]
```

```python
test_sents = brown_tagged_sents[size:]
```

```python
unigram_tagger = nltk.UnigramTagger(train_sents)
```

```python
unigram_tagger.evaluate(test_sents)
```

Although the score is worse, we now have **a better picture of the usefulness of this tagger**, i.e. **its performance on previously unseen text**.

## 5.3   General N-Gram Tagging

Unigram tagger tag a word such as *wind* with the same tag, **regardless of whether it appears in the context the wind or to wind**.

The best we can do is tag each word with its a priori most likely tag. (用先验最有可能的标签来标记每个单词)

An **n-gram tagger** is a generalization of a unigram tagger whose context is **the current word together with the part-of-speech tags of the n-1 preceding tokens**, as shown in below:

<div align=center>
<img src="https://www.nltk.org/images/tag-context.png">
<br>
<center><em><strong>Tagger Context</strong></em></center>
</div>

An n-gram tagger **picks the tag that is most likely in the given context**.

```python
bigram_tagger = nltk.BigramTagger(train_sents)
```

```python
print(bigram_tagger.tag(brown_sents[2007]))
```

```python
unseen_sent = brown_sents[4203]
```

```python
print(bigram_tagger.tag(unseen_sent))
```

Notice that the bigram tagger manages to tag every word in a sentence it saw during training, but does badly on an unseen sentence. (在训练数据中的句子上表现很好，因为在训练过程中见过该句子)

As soon as it encounters a new word (i.e., 13.5), it is unable to assign a tag. (当遇到训练数据中未见过的单词时，只能标记为None)

It cannot tag the following word (i.e., million) even if it was seen during training, simply because it never saw it during training with a None tag on the previous word. (None之后的词，也无法正确标注，因为其在训练数据中没见过这些单词的上一个词是None标签)

Consequently, the tagger fails to tag the rest of the sentence. Its overall accuracy score is very low.

```python
bigram_tagger.evaluate(test_sents)
```

As n gets larger, the specificity of the contexts increases, as does the chance that the data we wish to tag contains contexts that were not present in the training data. (随着n变大，上下文的特异性增加，我们希望标记的数据包含训练数据中不存在的上下文的机会也会增加)

This is known as the sparse data problem, and is quite pervasive in NLP. (这就是NLP中常见的数据稀疏问题) 

## 5.4   Combining Taggers

We could combine the results of a bigram tagger, a unigram tagger, and a default tagger, as follows:

1. Try tagging the token with the bigram tagger.

2. If the bigram tagger is unable to find a tag for the token, try the unigram tagger.

3. If the unigram tagger is also unable to find a tag, use a default tagger.

```python
t0 = nltk.DefaultTagger('NN')
t1 = nltk.UnigramTagger(train_sents, backoff=t0)
t2 = nltk.BigramTagger(train_sents, backoff=t1)
t2.evaluate(test_sents)
```

比仅使用unigram tagger准确性更好，提高了3%。
